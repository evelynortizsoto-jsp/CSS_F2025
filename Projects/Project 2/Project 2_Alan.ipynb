#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

The most pressing matter for feature selection was removing variables that were colinear with each other. Many variables were proportions of others or mutually exclusive, so were removed. For example, proportions of female/male in the total population, overall and by age. Thus, only female populations were kept, especially since earlier we had changed the variables to rates of populations within counties. Additionally, there were different kinds of age cutoffs, so the more granular age categories were kept. Finally, there was a lot of collinearity between the race alone *or* in combination with one or more race variables, which was unnecessary if we have other singular race variables. Identifying as Hispanic or Latino and not Hispanic or Latino and another racial identity was also extremely colinear with just identifying as that racial identity alone.

While some racial/ethnic identities represent a tiny fraction of the population (like Native and Pacific Islander populations), they were included in the dataset given the lack of representation these populations typically have within datasets. While it may be common practice to remove such sparse data, that practice has been perpetuated within the literature resulting in underrepresentation of such racial/ethnic groups and their health disaprities. Thus, we have elected to keep them in.

```python
#
# Validation split
#-----------

# train_test_split returns 4 values: X_train, X_test, y_train, y_test, so how do we create a 60-20-20 train-validate-test split? 
X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, train_size= 0.75, test_size= 0.25)    

print(f"Train: {len(X_train)} | Validate: {len(X_validate)} | Test: {len(X_test)}")
```

```python
#
# Standardization
#-----------
# Given that we want to only standardize non-dichotomous variables, we need to find a 
# solution that will loop over only the columns we want to standardize. The code below
# identifies all non-dichotomous variables in our dataset and only standardizes those.

# load library and create instance of Standard Scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


# identify non-dichotomous columns we want to transform
columns = list(X_test.select_dtypes(include=['number']).loc[:, X_test.nunique() > 2])

# use loop to transform training data for only columns we want to transform
for column in columns:
    X_train[column] = scaler.fit_transform(X_train[column].values.reshape(-1, 1)).flatten()

# use loop to transform validation data for only columns we want to transform
for column in columns:
    X_validate[column] = scaler.fit_transform(X_validate[column].values.reshape(-1, 1)).flatten()

# use loop to transform test data for only columns we want to transform
for column in columns:
    X_test[column] = scaler.fit_transform(X_test[column].values.reshape(-1, 1)).flatten()
```
